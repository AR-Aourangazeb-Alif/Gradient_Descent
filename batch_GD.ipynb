```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

# C L A S S E S

```python
class Batch_GD:
    def __init__(self):
        self.intercept = 1
        self.coef = None
        self.bita = None

    
    def fit(self, X, y, epochs, lr):
        X_dev = X.copy()
        X_dev.insert(0, 'intercept', 1)
        X_np = X_dev.values.astype(float)
        y_np = y.values.reshape(-1).astype(float)
        m = len(y_np)

        theta = np.zeros(X_np.shape[1])

        for _ in range(epochs):
            preds = X_np @ theta
            gradient = (2.0 / m) * (X_np.T @ (preds - y_np))
            theta = theta - lr * gradient

        self.bita = theta
        self.intercept = float(theta[0])
        self.coef = theta[1:]



    def predict(self, x):
        x_dev = x.copy()
        x_dev.insert(0,1,1)
        x_dev = np.array(x_dev)

        y = x_dev @ self.bita

        return y
```

```python
df_1 = pd.read_csv('Dhaka Rent.csv')
df_1.head()
```

```python
df_2 = pd.read_csv('Student_performance.csv')
df_2.drop('Extracurricular Activities', axis=1, inplace=True)
df_2.drop(df_2.index[:9000], inplace=True)
df_2
```

```python
x = df_2.iloc[:,:-1]
y = df_2.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1)
```

```python
test = Batch_GD()
test.fit(x_train, y_train, epochs=40, lr=0.01)
```

```python
test_pred = test.predict(x_test)
r2_score(y_test, test_pred)
```
